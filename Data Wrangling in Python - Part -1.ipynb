## Combining and Merging dataset

- pandas.merge
- pandas.concat
- combine_first

```python
df1 = pd.DataFrame({"key": ["b", "b", "a", "c", "a", "a", "b"],
                    "data1": pd.Series(range(7), dtype="Int64")})
df2 = pd.DataFrame({"key": ["a", "b", "d"],
                    "data2": pd.Series(range(3), dtype="Int64")})
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2)
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on='key')
```

```python
df3 = pd.DataFrame({"lkey": ["b", "b", "a", "c", "a", "a", "b"],
 ....: "data1": pd.Series(range(7), dtype="Int64")})
df4 = pd.DataFrame({"rkey": ["a", "b", "d"],
 ....: "data2": pd.Series(range(3), dtype="Int64")})
```

```python
df3
```

```python
df4
```

```python
pd.merge(df3,df4, left_on="lkey", right_on='rkey') # Column name are different
```

```python
pd.merge(df1, df2, how='outer')
```

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey', how = 'outer')
```

```python
df1 = pd.DataFrame({"key": ["b", "b", "a", "c", "a", "b"],
 ....: "data1": pd.Series(range(6), dtype="Int64")})
df2 = pd.DataFrame({"key": ["a", "b", "a", "b", "d"],
 ....: "data2": pd.Series(range(5), dtype="Int64")})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, how='left', on='key') #cartesian product
```

```python
pd.merge(df1, df2, how='inner')
```

```python
pd.merge(df1, df2, how='right')
```

```python
left = pd.DataFrame({"key1": ["foo", "foo", "bar"],
 ....: "key2": ["one", "two", "one"],
 ....: "lval": pd.Series([1, 2, 3], dtype='Int64')})
right = pd.DataFrame({"key1": ["foo", "foo", "bar", "bar"],
 ....: "key2": ["one", "one", "one", "two"],
 ....: "rval": pd.Series([4, 5, 6, 7], dtype='Int64')})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right)
```

```python
pd.merge(left, right, on = ['key1', 'key2'], how = 'outer')
```

```python
pd.merge(left, right, on="key2")
```

```python
#suffixes option for specifying strings to append to overlapping names in the left and right DataFrame
pd.merge(left, right, on="key1", suffixes=("_left", "_right"))
```

### Merging on Index


```python
left = pd.DataFrame({'key':["a", "b", "a", "a", "b", "c"], 
                 'values': range(6)})
left
```

```python
right = pd.DataFrame({"group_val": [3.5, 7]}, index=["a", "b"])
right
```

```python
pd.merge(left, right, left_on='key', right_index=True) #indicate that the index should be used as the merge key
```

```python
pd.merge(left, right, left_on='key', right_index=True, how='outer')
```

```python
lefth = pd.DataFrame({"key1": ["Ohio", "Ohio", "Ohio",
 ....: "Nevada", "Nevada"],
 ....: "key2": [2000, 2001, 2002, 2001, 2002],
 ....: "data": pd.Series(range(5), dtype="Int64")})
```

```python
righth_index = pd.MultiIndex.from_arrays( [
 ....: ["Nevada", "Nevada", "Ohio", "Ohio", "Ohio", "Ohio"],
 ....: [2001, 2000, 2000, 2000, 2001, 2002]
])
```

```python
lefth
```

```python
righth_index
```

```python
righth = pd.DataFrame({"event1":pd.Series([0, 2, 4, 6, 8, 10], index=righth_index),
                       'event2':pd.Series([1, 3, 5, 7, 9, 11], index=righth_index)
                      })
```

```python
righth
```

```python
pd.merge(lefth, righth, left_on=["key1", "key2"], right_index=True)
```

```python
pd.merge(lefth, righth, left_on=["key1", "key2"], right_index=True, how='outer')
```

```python
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],
 ....: index=["a", "c", "e"],
 ....: columns=["Ohio", "Nevada"]).astype("Int64")
```

```python
left2
```

```python
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
 ....: index=["b", "c", "d", "e"],
 ....: columns=["Missouri", "Alabama"]).astype("Int64")
```

```python
right2
```

```python
pd.merge(left2, right2, how='outer', left_index=True, right_index=True)
```

```python
left2.join(right2, how = 'outer')
```

```python
 another = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [16., 17.]],
 ....: index=["a", "c", "e", "f"],
 ....: columns=["New York", "Oregon"])
another
```

```python
left2.join([right2, another])
```

```python
left2.join([right2, another], how='outer')
```

### Concatenating Along an Axis

```python
arr = np.arange(12).reshape(3,4)
arr
```

```python
np.concatenate([arr, arr], axis=1)
```

```python
s1 = pd.Series([0, 1], index=["a", "b"], dtype="Int64")
s2 = pd.Series([2, 3, 4], index=["c", "d", "e"], dtype="Int64")
s3 = pd.Series([5, 6], index=["f", "g"], dtype="Int64")
```

```python
pd.concat([s1,s2,s3])
```

```python
pd.concat([s1,s2,s3], axis=1)
```

```python
result = pd.concat([s1, s2, s3], keys=["one", "two", "three"])
```

```python
result
```

```python
result.unstack()
```

```python
pd.concat([s1, s2, s3], axis=1,  keys=["one", "two", "three"])
```

```python
df1 = pd.DataFrame(np.arange(6).reshape(3, 2), index=["a", "b", "c"],
 .....: columns=["one", "two"])
df1
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis=1, keys=['level1','level2'])
```

```python
pd.concat([df1, df2], axis=1, keys=['level1','level2'], names=['upper','lower'])
```
