### Adjusting spaces around subplots

```python
fig, axes = plt.subplots(2,2, sharex=True, sharey=True)

for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.standard_normal(500), bins=20, color='black', alpha = 0.5)
fig.subplots_adjust(wspace=0, hspace=0)
```

### Color, Markup and Linestyle

```python
fig = plt.figure()
ax1 = fig.add_subplot()
ax1.plot(np.random.standard_normal(30).cumsum(), color = 'black', linestyle='dashed', marker='o');
```

```python
fig = plt.figure()
ax = fig.add_subplot()
data = np.random.standard_normal(30).cumsum()
ax.plot(data, color = 'black', linestyle='dashed', label = 'default')
ax.plot(data, color='blue', linestyle='dashed', drawstyle = 'steps-post', label = 'step-post');
```

### Adding legend

```python
fig, ax = plt.subplots()
ax.plot(np.random.randn(1000).cumsum(), color='black', linestyle = 'dashed', label='_nolegend_')
ax.plot(np.random.standard_normal(1000).cumsum(), color='blue', linestyle = 'dashed', label='two')
ax.plot(np.random.randn(1000).cumsum(), color='red', linestyle = 'dashed', label='three')
ax.legend()
```

### Annotation on a Subplot

```python
from datetime import datetime

fig , ax = plt.subplots()

data = pd.read_csv('spx.csv', index_col=0, parse_dates=True)
data.sort_index()
spx = data['SPX']
# spx = spx.sort_index()
# spx.sort_values(by=spx.index())
spx.plot(ax=ax, color='black')

crisis_data = [
    (datetime(2007,10,11), "Peak of Bull Market"),
    (datetime(2008,3,12), "Bear Stearns Fails"),
    (datetime(2008,9,15), "Lehman Bankruptcy")
]

for date, label in crisis_data:
    ax.annotate(label, xy = (date, spx.asof(date)+75), 
               xytext = (date, spx.asof(date)+225), 
                arrowprops=dict(facecolor = 'black', headwidth = 4, width=2, headlength = 4),
                horizontalalignment="left", verticalalignment="top")
    
ax.set_xlim(["1/1/2007", "1/1/2011"])
ax.set_ylim([600,1800]);
```
